```python
import polars as pl
import polars.selectors as cs
import duckdb as db
from great_tables import GT, loc, style, nanoplot_options
```

```python
df = pl.read_excel("Adidas US Sales Datasets.xlsx",read_options={"header_row": 3})
```

```python
%%markdown
## Retailer Report
```

```python
retailer_df = db.execute('SELECT CONCAT(Retailer,\'.png\') as icon, Retailer,ROUND(SUM("Total Sales"), 2) as sales FROM df GROUP BY 1,2 ORDER BY 3 DESC').pl()
```

```python
retailer_report = (GT(retailer_df)
                  .cols_label(sales="Sales", icon="")
                   .fmt_number(
                    columns=["sales"],
                    compact=True,
                    pattern="${x}",
                    n_sigfig=3)
                   .fmt_image("icon", path="./")
                  )
```

```python
retailer_df
```

```python
retailer_report
```

```python
%%markdown
## Region Report
```

```python
region_df = db.execute('SELECT Region,State,city, ROUND(SUM("Total Sales"), 2) as sales FROM df GROUP BY 1,2,3 ORDER BY 4 DESC').pl()
```

```python
region_df
```

```python
region_report = GT(region_df).tab_spanner(label="Total Sales", columns=cs.all()).fmt_number(
                    columns=["sales"],
                    compact=True,
                    pattern="${x}",
                    n_sigfig=3).data_color(
        columns=['sales'],
        palette=["#Dc863B", "#00A600", "#E6E600", "#E8C32E", "#D69C4E"]
        ).tab_stub(rowname_col="State", groupname_col="Region")
```

```python
region_report
```

```python
%%markdown
## Sales Method Analysis
```

```python
sales_method_df = db.execute('SELECT CONCAT("Sales Method", \'.png\') as icon,"Sales Method", ROUND(SUM("Total Sales"), 2) as sales FROM df GROUP BY 1,2 ORDER BY 3 DESC').pl()
```

```python
sales_method_df
```

```python
sales_method_report = GT(sales_method_df).tab_spanner(label="Total Sales By Method", columns=cs.all()).cols_label(sales="Sales", icon="").fmt_number(
                    columns=["sales"],
                    compact=True,
                    pattern="${x}",
                    n_sigfig=3).fmt_image("icon", path="./")
```
